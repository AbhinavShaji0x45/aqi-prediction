```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("clean_aqi.csv")

df['Date'] = pd.to_datetime(df['Date'])

plt.figure(figsize=(12,5))
plt.plot(df['Date'], df['AQI'])

plt.title("Chennai AQI Over Time")
plt.xlabel("Date")
plt.ylabel("AQI")

# FIX: reduce number of x-axis labels
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(10))

plt.tight_layout()
plt.show()
df['MA_7'] = df['AQI'].rolling(window=7).mean()

plt.figure(figsize=(12,5))
plt.plot(df['Date'], df['AQI'], alpha=0.5, label="Raw AQI")
plt.plot(df['Date'], df['MA_7'], color='red', label="7-Day Average")

plt.legend()
plt.title("AQI with Moving Average")
plt.show()

# FEATURE ENGINEERING FUNCTION (FIXES CODE DUPLICATION)
def create_features(df):
    """Create all features for AQI prediction with proper time series handling."""
    df = df.copy()
    
    # Lag features
    df['lag1'] = df['AQI'].shift(1)
    df['lag2'] = df['AQI'].shift(2)
    df['lag3'] = df['AQI'].shift(3)
    df['lag4'] = df['AQI'].shift(4)
    df['lag5'] = df['AQI'].shift(5)
    df['lag7'] = df['AQI'].shift(7)
    
    # Moving averages (using past values only - no leakage)
    df['ma7'] = df['AQI'].shift(1).rolling(7).mean()
    df['ma14'] = df['AQI'].shift(1).rolling(14).mean()
    df['ma30'] = df['AQI'].shift(1).rolling(30).mean()
    
    # Volatility features
    df['std7'] = df['AQI'].shift(1).rolling(7).std()
    df['std14'] = df['AQI'].shift(1).rolling(14).std()
    
    # Range features
    df['min7'] = df['AQI'].shift(1).rolling(7).min()
    df['max7'] = df['AQI'].shift(1).rolling(7).max()
    df['range7'] = df['max7'] - df['min7']
    
    # Exponential smoothing
    df['ema7'] = df['AQI'].shift(1).ewm(span=7, adjust=False).mean()
    
    # Temporal features
    df['day_of_week'] = df['Date'].dt.dayofweek
    df['month'] = df['Date'].dt.month
    df['quarter'] = df['Date'].dt.quarter
    df['day_of_year'] = df['Date'].dt.dayofyear
    
    return df.dropna()

print("✓ Feature engineering function created")
```

```python
# Time Series Cross-Validation (Walk-Forward)
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

df = pd.read_csv("clean_aqi.csv")

# Create lag features
df['lag1'] = df['AQI'].shift(1)
df['lag2'] = df['AQI'].shift(2)
df['lag3'] = df['AQI'].shift(3)
df = df.dropna()

X = df[['lag1', 'lag2', 'lag3']].values
y = df['AQI'].values

# Walk-forward validation
window_size = int(len(df) * 0.7)  # Initial training size
step = 50  # Test on 50 samples each fold

cv_scores = {'RMSE': [], 'MAE': [], 'R2': []}

for i in range(window_size, len(df) - step, step):
    # Train on data up to i
    X_train_wf = X[:i]
    y_train_wf = y[:i]
    
    # Test on next 'step' samples
    X_test_wf = X[i:i+step]
    y_test_wf = y[i:i+step]
    
    # Train model
    model = LinearRegression()
    model.fit(X_train_wf, y_train_wf)
    
    # Predict
    y_pred_wf = model.predict(X_test_wf)
    
    # Store metrics
    rmse = np.sqrt(mean_squared_error(y_test_wf, y_pred_wf))
    mae = mean_absolute_error(y_test_wf, y_pred_wf)
    r2 = r2_score(y_test_wf, y_pred_wf)
    
    cv_scores['RMSE'].append(rmse)
    cv_scores['MAE'].append(mae)
    cv_scores['R2'].append(r2)

print("=== Walk-Forward Cross-Validation Results ===\n")
print(f"Average RMSE: {np.mean(cv_scores['RMSE']):.4f} (+/- {np.std(cv_scores['RMSE']):.4f})")
print(f"Average MAE: {np.mean(cv_scores['MAE']):.4f} (+/- {np.std(cv_scores['MAE']):.4f})")
print(f"Average R²: {np.mean(cv_scores['R2']):.4f} (+/- {np.std(cv_scores['R2']):.4f})")

import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

axes[0].plot(cv_scores['RMSE'], marker='o')
axes[0].set_title('RMSE Across Folds')
axes[0].set_xlabel('Fold')
axes[0].set_ylabel('RMSE')
axes[0].grid()

axes[1].plot(cv_scores['MAE'], marker='o', color='orange')
axes[1].set_title('MAE Across Folds')
axes[1].set_xlabel('Fold')
axes[1].set_ylabel('MAE')
axes[1].grid()

axes[2].plot(cv_scores['R2'], marker='o', color='green')
axes[2].set_title('R² Across Folds')
axes[2].set_xlabel('Fold')
axes[2].set_ylabel('R²')
axes[2].grid()

plt.tight_layout()
plt.show()
```

```python
# Ensemble Methods - Combine Multiple Models
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, VotingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

df = pd.read_csv("clean_aqi.csv")

df['lag1'] = df['AQI'].shift(1)
df['lag2'] = df['AQI'].shift(2)
df['lag3'] = df['AQI'].shift(3)
df = df.dropna()

X = df[['lag1', 'lag2', 'lag3']]
y = df['AQI']

split = int(len(df) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Individual models
lr = LinearRegression()
ridge = Ridge(alpha=1.0)
rf = RandomForestRegressor(n_estimators=100, random_state=42, max_depth=10)
gb = GradientBoostingRegressor(n_estimators=100, random_state=42, max_depth=5)

# Voting Ensemble (average predictions)
voting_reg = VotingRegressor([
    ('lr', lr),
    ('ridge', ridge),
    ('rf', rf),
    ('gb', gb)
])

voting_reg.fit(X_train, y_train)
y_pred_voting = voting_reg.predict(X_test)

# Evaluate individual models
models = {
    'Linear Regression': lr,
    'Ridge': ridge,
    'Random Forest': rf,
    'Gradient Boosting': gb,
    'Voting Ensemble': voting_reg
}

results = {}
predictions = {}

for name, model in models.items():
    if name != 'Voting Ensemble':
        model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    predictions[name] = y_pred
    
    results[name] = {
        'RMSE': np.sqrt(mean_squared_error(y_test, y_pred)),
        'MAE': mean_absolute_error(y_test, y_pred),
        'R²': r2_score(y_test, y_pred)
    }

# Display results
print("=== Ensemble Model Comparison ===\n")
results_df = pd.DataFrame(results).T
print(results_df.to_string())

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Metrics comparison
metrics = ['RMSE', 'MAE', 'R²']
x = np.arange(len(models))
width = 0.25

for i, metric in enumerate(metrics):
    values = [results[model][metric] for model in models.keys()]
    axes[0].bar(x + i*width, values, width, label=metric, alpha=0.8)

axes[0].set_xlabel('Models')
axes[0].set_ylabel('Score')
axes[0].set_title('Model Performance Comparison')
axes[0].set_xticks(x + width)
axes[0].set_xticklabels(list(models.keys()), rotation=45, ha='right')
axes[0].legend()
axes[0].grid(axis='y', alpha=0.3)

# Plot 2: Predictions comparison
axes[1].plot(y_test.values, label='Actual', linewidth=2, marker='o', markersize=3)
for name, pred in predictions.items():
    axes[1].plot(pred, label=name, alpha=0.7, linestyle='--')

axes[1].set_xlabel('Sample')
axes[1].set_ylabel('AQI')
axes[1].set_title('Predictions Comparison (Ensemble vs Individual)')
axes[1].legend(fontsize=8, loc='best')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n✓ Voting Ensemble R² Score: {results['Voting Ensemble']['R²']:.4f}")
print(f"  Best individual: {max((v['R²'], k) for k, v in results.items() if k != 'Voting Ensemble')}")
```

```python
# ARIMA Time Series Model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_squared_error, mean_absolute_error
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv("clean_aqi.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Split data
train_size = int(len(df) * 0.8)
train = df['AQI'][:train_size]
test = df['AQI'][train_size:]

# Fit SARIMA model
# (p,d,q)(P,D,Q)m where m=7 for weekly seasonality
try:
    model = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 7))
    results = model.fit(disp=False)
    
    print("=== SARIMA Model Results ===\n")
    print(results.summary())
    
    # Forecast on test set
    forecast = results.get_forecast(steps=len(test))
    forecast_df = forecast.summary_frame()
    y_pred_arima = forecast_df['mean'].values
    
    # Evaluate
    rmse_arima = np.sqrt(mean_squared_error(test, y_pred_arima))
    mae_arima = mean_absolute_error(test, y_pred_arima)
    
    print(f"\n=== Test Set Performance ===")
    print(f"RMSE: {rmse_arima:.4f}")
    print(f"MAE: {mae_arima:.4f}")
    
    # Plot
    fig, axes = plt.subplots(2, 1, figsize=(12, 8))
    
    # Forecast plot
    axes[0].plot(train.index, train.values, label='Train', marker='o', markersize=2)
    axes[0].plot(test.index, test.values, label='Actual Test', marker='o', markersize=2)
    axes[0].plot(test.index, y_pred_arima, 'r--', label='SARIMA Forecast', linewidth=2)
    axes[0].fill_between(test.index, 
                          forecast_df['mean_ci_lower'].values, 
                          forecast_df['mean_ci_upper'].values, 
                          alpha=0.2, color='red')
    axes[0].set_title('SARIMA Model - AQI Forecast')
    axes[0].set_ylabel('AQI')
    axes[0].legend()
    axes[0].grid(alpha=0.3)
    
    # Residuals
    residuals = test.values - y_pred_arima
    axes[1].plot(residuals, marker='o', markersize=3)
    axes[1].axhline(y=0, color='r', linestyle='--')
    axes[1].set_title('SARIMA Residuals')
    axes[1].set_xlabel('Time')
    axes[1].set_ylabel('Residual')
    axes[1].grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
except Exception as e:
    print(f"ARIMA model encountered an issue: {e}")
    print("This may be due to data characteristics. Try with different (p,d,q) parameters.")
```

```python
# ADVANCED FEATURE ENGINEERING
import pandas as pd
import numpy as np

df = pd.read_csv("clean_aqi.csv")
df['Date'] = pd.to_datetime(df['Date'])

print("="*80)
print("ADVANCED FEATURE ENGINEERING".center(80))
print("="*80)

# 1. Extended Lag Features (lag1-7)
df['lag1'] = df['AQI'].shift(1)
df['lag2'] = df['AQI'].shift(2)
df['lag3'] = df['AQI'].shift(3)
df['lag4'] = df['AQI'].shift(4)
df['lag5'] = df['AQI'].shift(5)
df['lag7'] = df['AQI'].shift(7)

# 2. Moving Averages
df['ma7'] = df['AQI'].shift(1).rolling(window=7).mean()     # 7-day moving average
df['ma14'] = df['AQI'].shift(1).rolling(window=14).mean()   # 14-day moving average
df['ma30'] = df['AQI'].shift(1).rolling(window=30).mean()   # 30-day moving average

# 3. Standard Deviation (volatility)
df['std7'] = df['AQI'].shift(1).rolling(window=7).std()
df['std14'] = df['AQI'].shift(1).rolling(window=14).std()

# 4. Temporal Features
df['day_of_week'] = df['Date'].dt.dayofweek       # 0=Monday, 6=Sunday
df['month'] = df['Date'].dt.month                   # 1-12
df['quarter'] = df['Date'].dt.quarter               # 1-4
df['day_of_year'] = df['Date'].dt.dayofyear        # 1-365

# 5. Min/Max features
df['min7'] = df['AQI'].shift(1).rolling(window=7).min()
df['max7'] = df['AQI'].shift(1).rolling(window=7).max()
df['range7'] = df['max7'] - df['min7']

# 6. Exponential Moving Average
df['ema7'] = df['AQI'].shift(1).ewm(span=7, adjust=False).mean()

# Drop NaN values
df_clean = df.dropna()

# Display feature info
print(f"\nOriginal features: 3 (lag1, lag2, lag3)")
print(f"Enhanced features: {df_clean.shape[1] - 2} columns")
print(f"\nNew Features Added:")
feature_groups = {
    'Extended Lags': ['lag1', 'lag2', 'lag3', 'lag4', 'lag5', 'lag7'],
    'Moving Averages': ['ma7', 'ma14', 'ma30'],
    'Volatility': ['std7', 'std14'],
    'Temporal': ['day_of_week', 'month', 'quarter', 'day_of_year'],
    'Exponential Smoothing': ['ema7'],
    'Range': ['min7', 'max7', 'range7']
}

for group, features in feature_groups.items():
    print(f"  {group}: {', '.join(features)}")

print(f"\nFeature matrix shape: {df_clean.shape}")
print("\n✓ Feature engineering complete!")
```

```python
# ADVANCED ENSEMBLE & STACKING
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, VotingRegressor, StackingRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

print("="*80)
print("ADVANCED ENSEMBLE & STACKING".center(80))
print("="*80)

# Reload optimized data
df = pd.read_csv("clean_aqi.csv")
df['Date'] = pd.to_datetime(df['Date'])

df['lag1'] = df['AQI'].shift(1)
df['lag2'] = df['AQI'].shift(2)
df['lag3'] = df['AQI'].shift(3)
df['lag4'] = df['AQI'].shift(4)
df['lag5'] = df['AQI'].shift(5)
df['lag7'] = df['AQI'].shift(7)
df['ma7'] = df['AQI'].shift(1).rolling(window=7).mean()
df['ma14'] = df['AQI'].shift(1).rolling(window=14).mean()
df['ma30'] = df['AQI'].shift(1).rolling(window=30).mean()
df['std7'] = df['AQI'].shift(1).rolling(window=7).std()
df['std14'] = df['AQI'].shift(1).rolling(window=14).std()
df['day_of_week'] = df['Date'].dt.dayofweek
df['month'] = df['Date'].dt.month
df['quarter'] = df['Date'].dt.quarter
df['day_of_year'] = df['Date'].dt.dayofyear
df['min7'] = df['AQI'].shift(1).rolling(window=7).min()
df['max7'] = df['AQI'].shift(1).rolling(window=7).max()
df['range7'] = df['max7'] - df['min7']
df['ema7'] = df['AQI'].shift(1).ewm(span=7, adjust=False).mean()
df = df.dropna()

feature_cols = ['lag1', 'lag2', 'lag3', 'lag4', 'lag5', 'lag7', 
                'ma7', 'ma14', 'ma30', 'std7', 'std14',
                'day_of_week', 'month', 'quarter', 'day_of_year',
                'min7', 'max7', 'range7', 'ema7']

X = df[feature_cols]
y = df['AQI']

split = int(len(df) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

ensemble_results = {}

# 1. Weighted Voting Ensemble
print("\n--- Weighted Voting Ensemble ---")
base_models = [
    ('lr', LinearRegression()),
    ('rf', RandomForestRegressor(n_estimators=150, max_depth=12, random_state=42)),
    ('gb', GradientBoostingRegressor(n_estimators=150, max_depth=6, learning_rate=0.05, random_state=42))
]

voting_ensemble = VotingRegressor(estimators=base_models, weights=[0.2, 0.4, 0.4])
voting_ensemble.fit(X_train, y_train)
y_pred_voting = voting_ensemble.predict(X_test)

voting_r2 = r2_score(y_test, y_pred_voting)
voting_rmse = np.sqrt(mean_squared_error(y_test, y_pred_voting))
voting_mae = mean_absolute_error(y_test, y_pred_voting)

ensemble_results['Weighted Voting'] = {
    'R²': voting_r2,
    'RMSE': voting_rmse,
    'MAE': voting_mae
}

print(f"Weighted Voting R²: {voting_r2:.4f}")
print(f"Weighted Voting RMSE: {voting_rmse:.4f}")
print(f"Weighted Voting MAE: {voting_mae:.4f}")

# 2. Stacking Ensemble (Meta-learner approach)
print("\n--- Stacking Ensemble (with Ridge Meta-learner) ---")

base_learners = [
    ('rf', RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)),
    ('gb', GradientBoostingRegressor(n_estimators=100, max_depth=5, learning_rate=0.05, random_state=42)),
    ('ridge', Ridge(alpha=1.0))
]

meta_learner = Ridge(alpha=0.1)

# FIX: Use TimeSeriesSplit instead of random CV
stacking_ensemble = StackingRegressor(
    estimators=base_learners,
    final_estimator=meta_learner,
    cv=5  # Keep simple CV for stacking (TimeSeriesSplit causes issues with cross_val_predict)
)

stacking_ensemble.fit(X_train, y_train)
y_pred_stacking = stacking_ensemble.predict(X_test)

stacking_r2 = r2_score(y_test, y_pred_stacking)
stacking_rmse = np.sqrt(mean_squared_error(y_test, y_pred_stacking))
stacking_mae = mean_absolute_error(y_test, y_pred_stacking)

ensemble_results['Stacking'] = {
    'R²': stacking_r2,
    'RMSE': stacking_rmse,
    'MAE': stacking_mae
}

print(f"Stacking Ensemble R²: {stacking_r2:.4f}")
print(f"Stacking Ensemble RMSE: {stacking_rmse:.4f}")
print(f"Stacking Ensemble MAE: {stacking_mae:.4f}")

# Summary
print("\n" + "="*80)
print("ENSEMBLE RESULTS SUMMARY".center(80))
print("="*80)

ensemble_summary = pd.DataFrame(ensemble_results).T
print("\n" + ensemble_summary.to_string())

# Best ensemble
best_ensemble_name = ensemble_summary['R²'].idxmax()
best_ensemble_r2 = ensemble_summary['R²'].max()

print(f"\n✓ BEST ENSEMBLE: {best_ensemble_name}")
print(f"  R² Score: {best_ensemble_r2:.4f}")
print(f"  Explains {best_ensemble_r2*100:.2f}% of AQI variance")

# Visualization
fig, ax = plt.subplots(figsize=(10, 5))

methods = list(ensemble_results.keys())
r2_scores = [ensemble_results[m]['R²'] for m in methods]
colors = ['green' if r2 == max(r2_scores) else 'steelblue' for r2 in r2_scores]

bars = ax.bar(methods, r2_scores, color=colors, alpha=0.7, edgecolor='black')
ax.set_ylabel('R² Score', fontsize=12)
ax.set_title('Ensemble Methods Comparison', fontsize=14, fontweight='bold')
ax.set_ylim([0, 1])
ax.grid(axis='y', alpha=0.3)

# Add value labels on bars
for bar, score in zip(bars, r2_scores):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{score:.4f}',
            ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()

print("\n✓ Ensemble training complete!")
```
